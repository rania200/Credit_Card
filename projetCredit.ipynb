```python
import pandas as pd
import numpy as np

df = pd.read_csv("crx.data", header=None)
columns = ["Gender", "Age", "Debt", "Married", "BankCustomer", "EducationLevel", "Ethnicity", "YearsEmployed", "PriorDefault", "Employed", "CreditScore", "DriversLicense", "Citizen", "ZipCode", "Income", "ApprovalStatus"]
df.columns = columns

df.head()

df_description = df.describe()
print(df_description)

print("\n")


df_info = df.info()
print(df_info)

print("\n")


print(df.tail(17))


df = df.replace('?', np.nan)

df.tail(17)
```

```python
from IPython.display import display, HTML
title2 = '''
<h1>Data Preparation</h1>
<ul>
<li><h4>Handling the missing values</h4></li>
</ul>
<ul><h4>Cette partie du code se concentre sur le traitement des valeurs manquantes dans les données, qui peuvent fausser les prédictions d'approbation de carte de crédit. Voici les étapes clés :
</h4>
<h4 style=" color: green">1. Vérification du format des données:</h4>
<h4>- On s'assure que df est bien un objet de type DataFrame de Pandas, le format standard pour manipuler des données tabulaires.
</h4>
<h4 style=" color: green">2. Identification des colonnes numériques:</h4>
<h4>- On s'assure que df est bien un objet de type DataFrame de Pandas, le format standard pour manipuler des données tabulaires.
</h4>
<h4 style=" color: green">3. Imputation des valeurs manquantes numériques:</h4>
<h4> - Pour chaque colonne numérique, on calcule la moyenne. Cette valeur représente la tendance centrale des données.
 </h4><h4>- On remplace ensuite les valeurs manquantes dans ces colonnes par cette moyenne.
</h4>
<h4 style=" color: green">4. Vérification des valeurs manquantes restantes:</h4>
<h4>- On vérifie s'il reste des valeurs manquantes après avoir traité les données numériques.
</h4>
<h4 style=" color: green">5. Imputation des valeurs manquantes catégorielles:</h4>
<h4> - On passe en revue chaque colonne restante (généralement du texte comme l'état matrimonial).
 </h4><h4>- On identifie la valeur la plus fréquente (celle qui apparaît le plus souvent) dans chacune de ces colonnes.
 </h4><h4>- On remplace enfin les valeurs manquantes restantes par cette valeur la plus fréquente.
</h4>
<h4 style=" color: green">6. Vérification finale des valeurs manquantes:</h4>
<h4>- On effectue un dernier contrôle pour s'assurer qu'il ne reste plus aucune valeur manquante dans les données.
</h4>
<h4>En résumé, cette étape de préparation des données permet de "nettoyer" les informations en estimant de manière cohérente les valeurs manquantes. 
</h4><h4>C'est crucial pour que le modèle de prédiction d'approbation de carte de crédit puisse exploiter des données fiables et produire des résultats précis.
</h4></ul>'''
display(HTML(title2))
```

```python

# Ensure df is a DataFrame
df = pd.DataFrame(df)

# Select numeric columns
numeric_columns = df.select_dtypes(include=np.number)

# Impute missing values with mean for numeric columns
df[numeric_columns.columns] = df[numeric_columns.columns].fillna(df[numeric_columns.columns].mean())

# Check for any remaining missing values
print(df.isnull().sum())
```

```python
for col in df.columns:
    # Check if the column is of object type
    if df[col].dtypes == 'object':
        # Impute with the most frequent value
        df = df.fillna(df[col].value_counts().index[0])

# Count the number of NaNs in the dataset and print the counts to verify
print(df.isnull().sum())
```

```python
title2 = '''
<h1>Data Preprocessing</h1>

<h4>Cette partie du code prépare les données pour être utilisées efficacement par un modèle de machine learning afin de prédire les approbations de cartes de crédit.
</h4><h4>Elle comprend plusieurs étapes importantes :</h4>

<h4 style=" color: green">1-Encodage des variables catégorielles:</h4>
<h4> - Certaines colonnes, comme l'état matrimonial ou le pays de résidence, contiennent des valeurs textuelles.
 </h4><h4>- Le code utilise un LabelEncoder pour transformer ces valeurs en nombres entiers. Cela permet aux modèles de machine learning de comprendre et manipuler ces variables plus facilement.
</h4>
<h4 style=" color: green">2-Suppression de caractéristiques:</h4>
<h4>
 - Le code supprime les colonnes "CreditScore" et "Citizen" des données.
 </h4><h4>- Cette suppression peut être justifiée en fonction de l'analyse des données ou de l'expertise du domaine. Par exemple, le CreditScore pourrait déjà    être redondant avec d'autres informations financières, et la citoyenneté pourrait ne pas être un facteur pertinent pour l'approbation.

</h4>
<h4 style=" color: green">3-Séparation des caractéristiques et des labels:</h4>
<ul><li><h4> Les données sont divisées en deux parties :
    </h4><h4>- Les caractéristiques (X) représentent les informations sur les demandeurs (âge, revenu, etc.).
    </h4><h4>- Le label (y) représente la décision d'approbation (approuvé ou non).
</h4></ul></li>
<h4 style=" color:green">4-Division en ensembles d'entraînement et de test:</h4>
<h4>- Les données sont scindées en deux ensembles :
   </h4><h4>Ensemble d'entraînement (X_train, y_train) : utilisé pour entraîner le modèle de machine learning.
   </h4><h4>Ensemble de test (X_test, y_test) : utilisé pour évaluer la performance du modèle sur des données qu'il n'a pas vues auparavant. La taille habituelle de l'ensemble de test est de 20% à 30% des données totales.

<h4 style=" color: green">5-Mise à l'échelle des caractéristiques:</h4>
 <h4>- Le code utilise un MinMaxScaler pour mettre à l'échelle toutes les caractéristiques numériques entre 0 et 1.
 </h4><h4>- Cette mise à l'échelle est importante car certains modèles de machine learning sont sensibles à l'échelle des caractéristiques. Cela permet d'éviter  qu'une caractéristique n'ait plus de poids qu'une autre simplement parce qu'elle a des valeurs plus grandes.</h4>
<h4>En résumé, le prétraitement des données garantit que les informations sont présentées dans un format adapté aux modèles de machine learning, améliorant ainsi la prédiction des approbations de cartes de crédit.
</h4> '''

 
display(HTML(title2))
```

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le = LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in df.columns.to_numpy():
    # Compare if the dtype is object
    if df[col].dtypes =='object':
    # Use LabelEncoder to do the numeric transformation
        df[col]=le.fit_transform(df[col])
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Drop the features 11 and 13 and convert the DataFrame to a NumPy array
# df
df = df.drop(["CreditScore", "Citizen"], axis=1)
df = df.to_numpy()
#
#
# # Segregate features and labels into separate variables
X,y = df[:,1:13] , df[:,13]

#
# # Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size=0.33,
                                random_state=42)
X
```

```python
# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Instantiate MinMaxScaler and use it to rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

```python
title2 = '''
<h1>Modeling</h1>

<h4>Cette partie du code se concentre sur la construction d'un modèle de machine learning pour prédire les approbations de cartes de crédit.
</h4><h4>Elle utilise une régression logistique, une méthode statistique populaire pour les problèmes de classification binaire (approuvé ou non).</h4>
<h3 style=" color: blue">Étapes de la modélisation :</h3>
<h4 style=" color: green">1. Importation de la bibliothèque LogisticRegression:</h4>
<h4> 
 - La bibliothèque sklearn.linear_model est importée pour accéder à la classe LogisticRegression.</h4>
<h4 style=" color: green">2. Création d'un modèle de régression logistique:</h4>
<h4>
 
 - Une instance de LogisticRegression est créée avec les paramètres par défaut.
 </h4><h4>- Ces paramètres peuvent être personnalisés ultérieurement pour optimiser les performances du modèle.
</h4>
<h4 style=" color: green">3. Entraînement du modèle:</h4>
<h4>
 - La méthode fit() est utilisée pour entraîner le modèle sur les données d'entraînement (rescaledX_train) et les labels correspondants (y_train).
 </h4>
 <h4>- Au cours de l'entraînement, le modèle apprend à identifier les relations entre les caractéristiques (variables des demandeurs) et les labels (décisions d'approbation).
</h4>
<ul><li><h4> Rôle de la régression logistique:

 <h4>- La régression logistique est un algorithme d'apprentissage supervisé qui s'appuie sur une fonction logistique pour prédire la probabilité qu'une   observation appartienne à l'une des deux classes (approuvé ou non).
 </h4><h4>- En analysant les données d'entraînement, le modèle apprend à estimer les coefficients qui influencent le plus la probabilité d'approbation.
 </h4><h4>- Ces coefficients reflètent l'importance relative de chaque caractéristique dans la prise de décision d'approbation.
 

 </h4><h4>- Cette mise à l'échelle est importante car certains modèles de machine learning sont sensibles à l'échelle des caractéristiques. Cela permet d'éviter  qu'une caractéristique n'ait plus de poids qu'une autre simplement parce qu'elle a des valeurs plus grandes.</h4>
</li></ul>
<h4>En résumé, la phase de modélisation permet de créer un modèle statistique capable de prédire les approbations de cartes de crédit en fonction des caractéristiques des demandeurs.</h4> 
'''
display(HTML(title2))
```

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Instantiate a LogisticRegression classifier with default parameter values
logreg = LogisticRegression()

# Fit logreg to the train set
logreg.fit(rescaledX_train, y_train)
```

```python
# Import confusion_matrix
from sklearn.metrics import confusion_matrix

# Use logreg to predict instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test, y_test))

# Print the confusion matrix of the logreg model
print(confusion_matrix(y_test, y_pred))
```

```python
# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the grid of values for tol and max_iter
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200]

# Create a dictionary where tol and max_iter are keys and the lists of their values are corresponding values
param_grid = dict(tol= tol, max_iter= max_iter)
param_grid
```
